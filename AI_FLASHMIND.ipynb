````python
def generate_mcq_prompt(topic):
    return f"""
Generate 5 multiple-choice questions on the topic "{topic}".
Each question must have four options labeled A, B, C, D.
After each question, specify the correct option clearly.

Format:

1. Question: ...
   A) ...
   B) ...
   C) ...
   D) ...
   Correct Answer: B
"""
````

````python
import re

def parse_questions(response_text):
    pattern = r"\d+\.\s*Question:\s*(.*?)\n\s*A\)\s*(.*?)\n\s*B\)\s*(.*?)\n\s*C\)\s*(.*?)\n\s*D\)\s*(.*?)\n\s*Correct Answer:\s*([ABCD])"
    matches = re.findall(pattern, response_text, re.DOTALL)

    questions = []
    for match in matches:
        question, a, b, c, d, correct = match
        questions.append({
            "question": question.strip(),
            "options": [a.strip(), b.strip(), c.strip(), d.strip()],
            "answer": correct.strip()
        })
    return questions
````

````python
import ipywidgets as widgets
from IPython.display import display, clear_output
import time
import threading

def run_quiz(questions):
    score = 0
    total = len(questions)

    for i, q in enumerate(questions):
        clear_output(wait=True)
        print(f"🧠 Question {i+1} of {total}")
        print(q["question"])

        radio = widgets.RadioButtons(
            options=q["options"],
            layout=widgets.Layout(width='auto'),
            style={'description_width': 'initial'}
        )

        submit_button = widgets.Button(
            description="✅ Submit",
            button_style='success',
            layout=widgets.Layout(width='150px')
        )

        timer_output = widgets.Output()
        result_output = widgets.Output()
        box = widgets.VBox([radio, submit_button, timer_output, result_output])
        display(box)

        answered = threading.Event()

        def handle_submit(_):
            if not answered.is_set():
                selected = radio.value
                with result_output:
                    clear_output()
                    if selected and selected.strip().startswith(q["answer"]):
                        print("✅ Correct!")
                        nonlocal score
                        score += 1
                    else:
                        print(f"❌ Incorrect. Correct: {q['answer']}")
                answered.set()

        submit_button.on_click(handle_submit)

        def countdown():
            for t in range(10, 0, -1):
                with timer_output:
                    clear_output(wait=True)
                    print(f"⏳ Time left: {t} seconds")
                if answered.wait(1):
                    return
            if not answered.is_set():
                with result_output:
                    clear_output()
                    print(f"⏱️ Time's up! Correct: {q['answer']}")
                answered.set()

        timer_thread = threading.Thread(target=countdown)
        timer_thread.start()
        timer_thread.join()  # Wait for the timer to finish
        time.sleep(2)  # Optional delay before next question

    clear_output(wait=True)
    print(f"🏁 Quiz Completed! Your Score: {score}/{total}")
````

Room creating and joining

Let a host create a room (with a room code).

Let players join using that room code.

Show a list of players in the room.

````python
import random
import string
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- RoomManager Class ---
class RoomManager:
    def __init__(self):
        self.rooms = {}  # Format: {room_code: {players: []}}

    def generate_room_code(self, length=6):
        return ''.join(random.choices(string.ascii_uppercase + string.digits, k=length))

    def create_room(self):
        code = self.generate_room_code()
        self.rooms[code] = {"players": []}
        return code

    def join_room(self, code, player_name):
        if code in self.rooms:
            self.rooms[code]["players"].append(player_name)
            return True
        return False

    def get_players(self, code):
        return self.rooms.get(code, {}).get("players", [])

    def room_exists(self, code):
        return code in self.rooms

room_manager = RoomManager()

# --- UI Elements ---
create_button = widgets.Button(description="Create Room", button_style='success')
join_button = widgets.Button(description="Join Room", button_style='primary')
room_code_display = widgets.HTML()

name_input = widgets.Text(placeholder='Enter your name')
join_code_input = widgets.Text(placeholder='Enter room code to join')

player_list_output = widgets.Output()

# --- Functions ---
def handle_create_room(b):
    code = room_manager.create_room()
    room_code_display.value = f"<b>Room Created! Room Code: {code}</b>"
    join_code_input.value = code
    update_players(code)


def handle_join_room(b):
    name = name_input.value.strip()
    code = join_code_input.value.strip().upper()
    if not name or not code:
        room_code_display.value = "<b style='color:red;'>Enter both name and room code.</b>"
        return

    if room_manager.join_room(code, name):
        room_code_display.value = f"<b>✅ Joined Room {code} as {name}</b>"
        update_players(code)
    else:
        room_code_display.value = f"<b style='color:red;'>❌ Invalid Room Code: {code}</b>"


def update_players(code):
    with player_list_output:
        clear_output()
        players = room_manager.get_players(code)
        if players:
            print(f"👥 Players in Room {code}:")
            for p in players:
                print(f"• {p}")
        else:
            print("No players in this room yet.")

# --- Bind Buttons ---
create_button.on_click(handle_create_room)
join_button.on_click(handle_join_room)

# --- Layout ---
display(widgets.VBox([
    widgets.HTML("<h3>🎮 Multiplayer Quiz Room</h3>"),
    widgets.HBox([create_button, join_button]),
    widgets.HBox([name_input, join_code_input]),
    room_code_display,
    player_list_output
]))
````

````python
import random
import time
import threading
import ipywidgets as widgets
from IPython.display import display, clear_output

# Room & Quiz logic (same as before)
rooms = {}

def generate_quiz_questions(topic):
    return [
        {
            "question": "What is Photosynthesis?",
            "options": ["A) Process of making food", "B) Process of breathing", "C) Process of reproduction", "D) Process of digestion"],
            "answer": "A"
        },
        {
            "question": "What is the main product of Photosynthesis?",
            "options": ["A) Oxygen", "B) Glucose", "C) Carbon Dioxide", "D) Water"],
            "answer": "B"
        }
    ]

def generate_room_code():
    return ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=6))

def create_room(topic):
    room_code = generate_room_code()
    rooms[room_code] = {'topic': topic, 'players': []}
    return room_code

def join_room(room_code, player_name):
    if room_code not in rooms:
        print("❌ Invalid room code.")
        return

    if 'questions' not in rooms[room_code]:
        rooms[room_code]['questions'] = generate_quiz_questions(rooms[room_code]['topic'])

    rooms[room_code]['players'].append({
        'name': player_name,
        'questions': rooms[room_code]['questions'],
        'score': 0,
        'current_question_index': 0,
        'answers': []  # Track user's selected answers
    })

    print(f"✅ {player_name} joined room: {room_code}")

# Updated PlayerQuiz class
class PlayerQuiz:
    def __init__(self, room_code, player_name, time_limit=60):
        self.room_code = room_code
        self.player_name = player_name
        self.time_limit = time_limit
        self.player = next(p for p in rooms[room_code]['players'] if p['name'] == player_name)
        self.answer_dropdown = None
        self.submit_button = None
        self.timer_label = widgets.Label()
        self.quiz_box = widgets.VBox()
        self.end_time = None
        self.timer_thread = None
        self.quiz_ended = False

    def load_question(self):
        clear_output(wait=True)
        current_index = self.player['current_question_index']

        if current_index >= len(self.player['questions']):
            self.quiz_ended = True
            self.show_results()
            return

        q = self.player['questions'][current_index]
        print(f"Question {current_index + 1}: {q['question']}")

        self.answer_dropdown = widgets.Dropdown(
            options=q['options'],
            description='Answer:',
            layout=widgets.Layout(width='70%')
        )
        self.submit_button = widgets.Button(description="Submit ✅", button_style='success')
        self.submit_button.on_click(self.submit_answer)

        self.quiz_box.children = [self.timer_label, self.answer_dropdown, self.submit_button]
        display(self.quiz_box)

    def submit_answer(self, b):
        selected_option = self.answer_dropdown.value
        correct_answer = self.player['questions'][self.player['current_question_index']]['answer']

        self.player['answers'].append(selected_option)

        if selected_option.startswith(correct_answer):
            self.player['score'] += 1

        self.player['current_question_index'] += 1
        self.load_question()

    def start_timer(self):
        while time.time() < self.end_time and not self.quiz_ended:
            remaining = int(self.end_time - time.time())
            self.timer_label.value = f"⏳ Time Left: {remaining} seconds"
            time.sleep(1)

        if not self.quiz_ended:
            self.quiz_ended = True
            self.show_results()

    def show_results(self):
        clear_output(wait=True)
        print(f"🏁 Time's up or Quiz Completed!\n✅ Your Score: {self.player['score']}/{len(self.player['questions'])}")
        print("\n📋 Here's what you answered:")

        for idx, q in enumerate(self.player['questions']):
            correct = q['answer']
            user_ans = self.player['answers'][idx] if idx < len(self.player['answers']) else "No answer"
            result = "✅ Correct" if user_ans.startswith(correct) else "❌ Incorrect"
            print(f"\nQ{idx+1}: {q['question']}")
            print(f"Your Answer: {user_ans}")
            print(f"Correct Answer: {q['options'][ord(correct)-65]}")
            print(f"Result: {result}")

    def start(self):
        print(f"⏳ Starting quiz for {self.player_name}! You have {self.time_limit} seconds.")
        self.end_time = time.time() + self.time_limit

        # Start timer thread
        self.timer_thread = threading.Thread(target=self.start_timer)
        self.timer_thread.start()

        self.load_question()
````

````python
import ast

def generate_quiz_questions(topic):
    prompt = f"""
    Generate exactly 5 multiple-choice questions on the topic "{topic}".
    Format each question as a Python dictionary like this:
    {{
      "question": "Your question here?",
      "options": ["A) Option 1", "B) Option 2", "C) Option 3", "D) Option 4"],
      "answer": "B"
    }}
    Return ONLY a Python list of 5 such dictionaries. No markdown, no explanation, just pure Python list.
    """

    try:
        model = genai.GenerativeModel("models/gemini-1.5-pro")
        response = model.generate_content(prompt)
        raw_text = response.text.strip()

        # Clean the response to remove code block markers
        if raw_text.startswith('```python') and raw_text.endswith('```'):
            raw_text = raw_text[9:-3].strip()  # Remove the ```python and ```

        # Attempt to parse using ast.literal_eval
        data = ast.literal_eval(raw_text)

        # Validate structure
        valid_questions = []
        for q in data:
            if (
                isinstance(q, dict)
                and 'question' in q
                and 'options' in q and len(q['options']) == 4
                and 'answer' in q and q['answer'] in ['A', 'B', 'C', 'D']
            ):
                valid_questions.append(q)

        if len(valid_questions) < 5:
            print(f"⚠️ Only {len(valid_questions)} valid questions parsed.")
        return valid_questions

    except Exception as e:
        print("❌ Error parsing Gemini response:", e)
        print("Raw response was:", raw_text)
        return []
````
